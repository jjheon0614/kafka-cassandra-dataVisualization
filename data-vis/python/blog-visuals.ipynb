```python
vancouver_weather = weather[weather['location'] == 'Vancouver']
seoul_weather = weather[weather['location'] == 'Seoul']
sydney_weather = weather[weather['location'] == 'Sydney']
```

```python
plt.figure(figsize=(20,6))
plt.plot(vancouver_weather['forecast_timestamp'], vancouver_weather['temp'] - 273.15)
plt.xlabel("Time")
plt.ylabel("Temperature")
plt.title("Vancouver weather data recorded")
```

```python
btc_table = crypto[['btc', 'date']].copy()

# Sort the DataFrame by the 'date' column in ascending order
btc_table = btc_table.sort_values(by='date')


eth_table = crypto[['eth', 'date']].copy()

# Sort the DataFrame by the 'date' column in ascending order
eth_table = eth_table.sort_values(by='date')

doge_table = crypto[['doge', 'date']].copy()

# Sort the DataFrame by the 'date' column in ascending order
doge_table = doge_table.sort_values(by='date')
```

```python
# Your existing code for plotting btc data
plt.figure(figsize=(20, 6))
plt.plot(btc_table['date'], btc_table['btc'])
plt.xlabel("Time")
plt.ylabel("btc")
plt.title("btc data recorded")

# Set the number of ticks you want on the y-axis
num_ticks = 5

# Manually set y-ticks
plt.yticks(btc_table['btc'][::len(btc_table['btc']) // (num_ticks - 1)].tolist())

# Show the plot
plt.show()
```
